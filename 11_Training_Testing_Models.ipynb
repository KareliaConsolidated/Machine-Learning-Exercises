### Training and Testing Models

```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=2)

# Load Data
loc = pd.read_csv('Datasets/LinearRegression/train_data.csv')
loc.head()
```

```python
# Split the Data
X = np.array(loc[['x1','x2']])
y = np.array(loc['y'])
```

```python
# Import Statements for the Classfication Algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
# Logistic Regression Classifier
logclassifier = LogisticRegression(solver='lbfgs')
logclassifier.fit(X,y)


# # DecisionTreeClassifier
dtclassifier = DecisionTreeClassifier()
dtclassifier.fit(X,y)

# # Support Vector Machine Classifier
svmclassifier = SVC(gamma='auto')
svmclassifier.fit(X,y)
```

```python
def plot_model(X,y,clf,title):
    
    plt.scatter(X[np.argwhere(y==0).flatten(),0],X[np.argwhere(y==0).flatten(),1], s=50, color='blue', alpha=0.75, edgecolor='k')
    plt.scatter(X[np.argwhere(y==1).flatten(),0],X[np.argwhere(y==1).flatten(),1], s=50, color='red', alpha=0.75, edgecolor='k')
    
    plt.xlim(-2.05, 2.05)
    plt.ylim(-2.05, 2.05)
    plt.grid(False)
    plt.tick_params(axis='x', which='both', bottom='off', top='off')
    plt.title(title)
    r = np.linspace(-2.1,2.1,300)
    s,t = np.meshgrid(r,r)
    s = np.reshape(s,(np.size(s),1))
    t = np.reshape(t,(np.size(t),1))
    h = np.concatenate((s,t),1)

    z = clf.predict(h)

    s = s.reshape((np.size(r),np.size(r)))
    t = t.reshape((np.size(r),np.size(r)))
    z = z.reshape((np.size(r),np.size(r)))

    plt.contourf(s,t,z,colors = ['blue','red'],alpha = 0.2,levels = range(-1,2))
    if len(np.unique(z)) > 1:
        plt.contour(s,t,z,colors = 'k', linewidths = 2)
    plt.show()
    
plot_model(X,y,logclassifier,'Logistic Classifier')
plot_model(X,y,dtclassifier,'Decision Tree Classifier')
plot_model(X,y,svmclassifier,'Support Vector Machine Classifier')
```

It seems that Logistic Regression didn't do so well, as it's a linear algorithm. Decision Trees managed to bound the data well
#### Tuning Parameters

```python
# # Support Vector Machine Classifier
svmclassifiertunned = SVC(kernel = 'rbf', gamma = 200)
svmclassifiertunned.fit(X,y)
plot_model(X,y,svmclassifiertunned,'Support Vector Machine Classifier')
```

#### Testing Model

```python
# Import Statements
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Read the Data
data = np.asarray(pd.read_csv('Datasets/LinearRegression/train_test_data.csv', header=None))

# Assign the feature to the variable X, and the labels to the variable y
X = data[:,0:2]
y = data[:,2]

# Use a train test split to split your data
# Use a test size of 25% and random state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create the decision tree model and assign it to the variable model
model = DecisionTreeClassifier()

# Fit the model to the training data
model.fit(X_train,y_train)

# Make Predictions on the test data
y_pred = model.predict(X_test)

# Calculate the accuracy and assign it to the variable acc on the test data
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {round(acc * 100)}%")

def load_pts(X,y,clf,title):
    
    plt.scatter(X[np.argwhere(y==0).flatten(),0],X[np.argwhere(y==0).flatten(),1], s=50, color='blue', alpha=0.75, edgecolor='k')
    plt.scatter(X[np.argwhere(y==1).flatten(),0],X[np.argwhere(y==1).flatten(),1], s=50, color='red', alpha=0.75, edgecolor='k')
    
    plt.xlim(0, 1.0)
    plt.ylim(0, 1.0)
    plt.grid(False)
    plt.tick_params(axis='x', which='both', bottom='off', top='off')
    plt.title(title)
    r = np.linspace(-2.1,2.1,300)
    s,t = np.meshgrid(r,r)
    s = np.reshape(s,(np.size(s),1))
    t = np.reshape(t,(np.size(t),1))
    h = np.concatenate((s,t),1)

    z = clf.predict(h)

    s = s.reshape((np.size(r),np.size(r)))
    t = t.reshape((np.size(r),np.size(r)))
    z = z.reshape((np.size(r),np.size(r)))

    plt.contourf(s,t,z,colors = ['blue','red'],alpha = 0.2,levels = range(-1,2))
    if len(np.unique(z)) > 1:
        plt.contour(s,t,z,colors = 'k', linewidths = 0.5)
    
    plt.show()
    
load_pts(X,y,model,'DecisionTreeClassifier')
```
